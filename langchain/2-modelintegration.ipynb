# Streaming
Most models can stream their output while it is being generated

```python
for chunk in model.stream("Why do parrots talk?"):
    print(chunk.text,end="|",flush=True)
```

# Batch 
### It is a technique where independent request to a model can significantly reduce cost and improve performance as the processing can be done in parrallel

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="qwen/qwen3-32b")
responses=model.batch(
    ["Why do parrots have colorful feathers?","How do airplanes fly?","What is quantum computing?"],
    config={
        'max_concurrency':5 # Limit to 5 parrallel calls
    }
)
for response in responses:
    print(response)
```
