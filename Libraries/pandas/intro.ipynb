## pd.read_csv(): 
- Best for simple, human-readable, and lightweight data. Slower parsing, larger file sizes, and prone to losing data types (e.g., stripping leading zeros)

## pd.read_excel():
- Designed for working directly with Excel spreadsheets. It is the slowest of the three and may choke on files exceeding 100,000 rows.

## pd.read_parquet():
- The gold standard for data science and large datasets. It uses a binary columnar format, offering lightning-fast read speeds, up to 75% smaller file sizes, and strict data-type preservation
